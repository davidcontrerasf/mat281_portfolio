3. Desarrolle una función `resumen_df(df)` para encontrar el total de elementos distintos y vacíos por columnas.

```python
def resumen_df(df):
    """
    funcion resumen con elementos distintos y vacios
    por columnas
    """
    nombres = df.columns

    u_list=[] # lista con elementos únicos
    n_list=[] # lista con elementos vacíos

    for j in nombres:

      unique=(len(df[j].unique()))
      vacio=(df[j].isnull().sum())

      if vacio!=0:
         unique=unique-1;

      u_list.append(unique) # agrega la cantidad de elementos distintos que hay por columna a la lista
      n_list.append(vacio) # agrega la cantidad de elementos vacíos que hay por columna a la lista

    result = pd.DataFrame({'nombres': nombres})
    result['elementos_distintos'] = u_list #ponemos cada componente de la lista unique en su correspondiente lugar en el df result
    result['elementos_vacios'] = n_list #ponemos cada componente de la lista null en su correspondiente lugar en el df result

    return result
```

```python
# retornar
resumen_df(df)
```

4. Para los paises latinoamericano, encuentre por año  el país con mayor y menor `indice`.

 * a) Mediante un ciclo _for_.
 * b) Mediante un  _groupby_.

```python
# Solucion

america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
       'CUB', 'DOM', 'ECU', 'GRD', 'GTM', 'GUY', 'HND', 'HTI', 'JAM',
       'MEX', 'NIC', 'PAN', 'PER', 'PRY', 'SLV', 'SUR', 'TTO', 'URY',
       'USA', 'VEN']

df_america = df[df.codigo_iso.isin(america)]

# Se creará un diccionario con los países con mayor y menor índice
# maximo

anios = df['anio'].unique() # lista de años

dict_mayor= dict()

for k in anios:
    df_aux = df_america.loc[lambda x: x['anio'] == k] # se filtra por cada año
    max_indice = df_aux['indice'].max()  # máximo de la columna
    df_aux = df_aux[df_aux['indice']==max_indice]
    paises = list(df_aux['pais'])

    new_paises=[]

    if len(paises)<1:
      paises.append('No hay datos')
      new_paises.append(paises[0])
    elif len(paises)>=1:
      new_paises.append(paises[0])
    if (np.isnan(max_indice)==True): # si el valor es NaN se cambia por un 0
      max_indice=0

    dict_mayor[k]=[new_paises[0],max_indice] # tupla con el nombre del país y su índice en el diccionario

sorted(dict_mayor.items()) # ordena por año
```

```python
# Solución

# minimo
dict_menor= dict()

for k in anios:
    df_aux = df_america.loc[lambda x: x['anio'] == k] # se filtra por cada año
    min_indice = df_aux['indice'].min()  # encontrar mínimo
    df_aux=df_aux[df_aux['indice']==min_indice]
    paises=list(df_aux['pais'])

    new_paises=[]

    if len(paises)<1:
      paises.append('No hay datos')
      new_paises.append(paises[0])
    elif len(paises)>=1:
      new_paises.append(paises[0])
    if (np.isnan(min_indice)==True): #si el valor es NaN se cambia por un 0
      min_indice=0
    dict_menor[k]=[new_paises[0],min_indice]

sorted(dict_menor.items()) # ordena por año
```

```python
# maximo

max_indice = df_america.groupby('anio').indice.max() # se asigna a cada año el mayor indice por país
data_max = df_america.merge(max_indice, on='anio', suffixes=('','_max')) # dataframe ordenado por año
data_max= data_max[data_max.indice==data_max.indice_max].drop(['indice_max','codigo_iso','ranking',], axis=1)
by_anio = data_max.sort_values('anio') # se ordena por anio

by_anio
```

```python
#b)minimo

min_indice = df_america.groupby('anio').indice.min() # se asigna a cada año el menor indice por país
data_min = df_america.merge(min_indice, on='anio', suffixes=('','_min')) # dataframe ordenado
data_min= data_min[data_min.indice==data_min.indice_min].drop(['indice_min','codigo_iso','ranking',], axis=1)
by_anio = data_min.sort_values('anio')

by_anio
```
